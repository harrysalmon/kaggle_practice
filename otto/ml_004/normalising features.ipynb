```python
import numpy as np
import pandas as pd 
```

```python
# load training data
X_y_train = np.genfromtxt('/home/harry/Documents/git/kaggle/otto/ml_004/X_y_train.csv', delimiter=',', dtype='float32')
X_train = X_y_train[:,0:93]
y_train = X_y_train[:,93]
# load cross validation data
X_y_cv = np.genfromtxt('/home/harry/Documents/git/kaggle/otto/ml_004/X_y_cv.csv', delimiter=',', dtype='float32')
X_cv = X_y_cv[:,0:93]
y_cv = X_y_cv[:,93]
# load test data
X_test = np.genfromtxt('/home/harry/Documents/git/kaggle/otto/ml_004/test_X_y.csv', delimiter=',', dtype='float32')
X_test = X_test[:,0:93]
```

```python
means = np.mean(X_train,0) # columns mean
means_array = means*np.ones(shape=(np.shape(X_train)[0],1))
stds = np.std(X_train,0) # columns std
stds_array = stds*np.ones(shape=(np.shape(X_train)[0],1))
X_train_check = (X_train - means_array) / stds_array
print X_train_check[0:2,:]
```

```python
means_array_cv = means*np.ones(shape=(np.shape(X_cv)[0],1))
stds_array_cv = stds*np.ones(shape=(np.shape(X_cv)[0],1))
X_cv_scaled = (X_cv - means_array_cv) / stds_array_cv
print X_cv_scaled[0:2,:]
np.savetxt('/home/harry/Documents/git/kaggle/otto/ml_004/X_cv_scaled.csv', X_cv_scaled, delimiter=',')
```

```python
means_array_test = means*np.ones(shape=(np.shape(X_test)[0],1))
stds_array_test = stds*np.ones(shape=(np.shape(X_test)[0],1))
X_test_scaled = (X_test - means_array_test) / stds_array_test
print X_test_scaled[0:2,:]
np.savetxt('/home/harry/Documents/git/kaggle/otto/ml_004/X_test_scaled.csv', X_test_scaled, delimiter=',')
```
